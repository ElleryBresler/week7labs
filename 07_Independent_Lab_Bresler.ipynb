#### Merge the Data 

```python
hospital_merge = ca_hospital_data.merge(ca_hospital_personnel, how='left', on='HospitalID')

hospital_merge.info() #check to make sure it worked 
```

```python
hospital_merge = hospital_merge.drop(['Work_ID', 'PositionID','Website'], axis = 1)
```

#### Exporting the Data

```python
#filtering out the data
hospital_data_new = hospital_merge[
    (hospital_merge['Teaching'] == 'Small/Rural') &
    (hospital_merge['AvlBeds'] >= 15) &
    (hospital_merge['OperInc'] >= 0)
]
```
